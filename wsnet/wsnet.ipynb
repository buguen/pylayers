```python
import os 
file_shadowing = os.listdir('./ban-propag/csv/')
```

```python
import csv

d = {}
for fn in file_shadowing:
    with open('./ban-propag/csv/'+fn,'rb') as csvfile:
        sha_reader=csv.reader(csvfile,delimiter=',')
        sh = sha_reader.next()
        t = []
        for k in sh:
            t .append(eval(k))
        d[fn.replace('shadowing-','').replace('.csv','')]=t    
```

```python
figsize(10,10)
for k in d:
    plot(d[k])
legend(d.keys())
title ('shadowing as a function of time')
```
